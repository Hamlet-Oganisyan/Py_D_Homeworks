```python
class Student:
    def __init__(self, name, surname, gender):
        self.name = name
        self.surname = surname
        self.gender = gender
        self.finished_courses = []
        self.courses_in_progress = []
        self.grades = {}
 
    def add_courses(self, course_name):
        self.finished_course.append(course_name)   
 
    def rate_lecturers(self, lecturer, course, grade):
        if isinstance(lecturer, Lecturer) and course in self.courses_in_progress and course in lecturer.courses_attached:
            if course in lecturer.grades:
                lecturer.grades[course] += [grade]
            else:
                lecturer.grades[course] = [grade]
        else:
            return 'Ошибка'
        
    def avg_grades(self):
        for self.avg_grade in self.grades.values():
            self.avg_grade = round(sum(self.avg_grade)/len(self.avg_grade), 1)
            return self.avg_grade   
    
    def __str__ (self):
        return f' Имя: {self.name} \n Фамилия: {self.surname} \n Средняя оценка за домашнее задание: {self.avg_grade} \
    \n курсы в процкссе изучения: {self.courses_in_progress} \n Завершенные курсы: {self.finished_courses}'
    
    def __lt__(self, other):
        if not isinstance(other, Student):
            print('Это не студент!')
            return
        return self.avg_grade < other.avg_grade
        
     
class Mentor:
    def __init__(self, name, surname):
        self.name = name
        self.surname = surname
        self.courses_attached = []
 
class Lecturer(Mentor):
    def __init__(self,name, surname):
        super().__init__(name, surname)
        self.grades = {}
    
    def avg_grades(self):
        for self.avg_grade in self.grades.values():
            self.avg_grade = round(sum(self.avg_grade)/len(self.avg_grade), 1)
            return self.avg_grade   
            
    def __str__ (self):
        return f' Имя: {self.name} \n Фамилия: {self.surname} \n Средняя оценка за лекции: {self.avg_grade}'
    
    def __lt__(self, other):
        if not isinstance(other, Lecturer):
            print('Это не лектор!')
            return
        return self.avg_grade < other.avg_grade
            
class Reviewer(Mentor):
    def rate_hw(self, student, course, grade):
        if isinstance(student, Student) and course in self.courses_attached and course in student.courses_in_progress:
            if course in student.grades:
                student.grades[course] += [grade]
            else:
                student.grades[course] = [grade]
        else:
            return 'Ошибка'
    
    def __str__ (self):
        return f' Имя: {self.name} \n Фамилия: {self.surname}'
    
# Создадим по 2 экземпляра класса Student
student_1 = Student('Александр', 'Волков', 'Муж')
student_1.courses_in_progress = ['Python', 'Git']
student_1.finished_courses = ['Введение в программирование']
student_2 = Student('Ирина', 'Петрова', 'Жен')
student_2.courses_in_progress = ['Python', 'Git', 'English for programmers']
student_2.finished_courses= ['Введение в программирование']

# Создадим по 2 экземпляра класса Lecturer
lecturer_1 = Lecturer('Виктор', 'Жуков')
lecturer_1.courses_attached += ['Python']
lecturer_2 = Lecturer('Андрей', 'Попов')
lecturer_2.courses_attached += ['Git']

# Создадим по 2 экземпляра класса Reviewer
reviewer_1 = Reviewer('Оксана', 'Сергеева')
reviewer_1.courses_attached += ['Python']
reviewer_2 = Reviewer('Виолетта', 'Дубинина')
reviewer_2.courses_attached += ['Python', 'Git']

# Реализуем функцию для подсчета средних оценок лекторов и студентов
def avg(list, name):
    list_1 = []
    list_2 = []
    for n in list:
        if name in n.keys():
            for new in n[name]:
                list_1.append(new)
            list_2 = round(sum(list_1)/len(list_1), 1)
    return list_2

### Проставим оценки студентам нашими ревьюверами
reviewer_1.rate_hw(student_1, 'Python', 9)
reviewer_1.rate_hw(student_1, 'Git', 7 )
reviewer_1.rate_hw(student_2, 'Python', 8)
reviewer_1.rate_hw(student_2, 'Git', 8 )
reviewer_2.rate_hw(student_1, 'Python', 7)
reviewer_2.rate_hw(student_1, 'Git', 7)
reviewer_2.rate_hw(student_2, 'Python', 6)
reviewer_2.rate_hw(student_2, 'Git', 8)
total_student_avg_list = [student_1.grades, student_2.grades]

### Проставим оценки лекторам нашими студентами
student_1.rate_lecturers(lecturer_1, 'Python', 8)
student_1.rate_lecturers(lecturer_2, 'Git', 7)
student_2.rate_lecturers(lecturer_1, 'Python', 7)
student_2.rate_lecturers(lecturer_2, 'Git', 6)
student_1.rate_lecturers(lecturer_1, 'Python', 7)
student_1.rate_lecturers(lecturer_2, 'Git', 7)
student_2.rate_lecturers(lecturer_1, 'Python', 5)
student_2.rate_lecturers(lecturer_2, 'Git', 6)
total_lecturer_avg_list = [lecturer_1.grades, lecturer_2.grades]

student_1.avg_grades()
student_2.avg_grades()
lecturer_1.avg_grades()
lecturer_2.avg_grades()

print('\033[1m' + 'Ревьюверы: ' + '\033[0m')
print(reviewer_1)
print(reviewer_2)
print('\033[1m' + '-'*60 + '\033[0m')
print('\033[1m' + 'Лекторы: ' + '\033[0m')
print(lecturer_1)
print()
print(lecturer_2)
print()
print(f'Средняя оценка всех лекторов за все курсы Python: {avg(total_lecturer_avg_list, "Python")}')
print(f'Средняя оценка всех лекторов за все курсы GIT: {avg(total_lecturer_avg_list, "Git")}')
print('\033[1m' + '-'*60 + '\033[0m')
print('\033[1m' + 'Студенты: ' + '\033[0m')  
print(student_1)
print()
print(student_2)
print()
print(f'Средняя оценка всех студентов за все курсы Git: {avg(total_student_avg_list, "Git")}')
print(f'Средняя оценка всех студентов за все курсы Python: {avg(total_student_avg_list, "Python")}')
```
